# **Ejemplo 1**

Se tienen dos cajas, una roja, con dos manzanas y seis fresas, y otra azul, con tres manzanas y una fresa:

*   Variables: $C = \{r{\rm oja},a{\rm zul}\}$ y $F = \{f{\rm resa},m{\rm anzana}\}$.

Al realizar $n=10$ experimentos aleatorios e independientes, las probabilidades de seleccionar una caja roja y una caja azul son:

*   Experimentos: $p(C=r)=4/10$ y $p(C=a)=6/10$.

Calcular las siguientes probabilidades:

1.   Las probabilidades condicionales de seleccionar un tipo de fruta dado un color de caja seleccionada:

  *   La probabilidad de seleccionar una manzana dada una caja roja: $p(F=m|C=r)=1/4$.
  *   La probabilidad de seleccionar una fresa dada una caja roja: $p(F=f|C=r)=3/4$.
  *   La probabilidad de seleccionar una manzana dada una caja azul: $p(F=m|C=a)=3/4$.
  *   La probabilidad de seleccionar una fresa dada una caja azul: $p(F=f|C=a)=1/4$.





2. La probabilidad total de seleccionar una manzana equivale a la probabilidad marginal:
\begin{eqnarray*}
p(F=m)  &  = & p(F=m|C=r)p(C=r) + p(F=m|C=a)p(C=a), \\
               & = & \frac{1}{4} \times \frac{4}{{10}} + \frac{3}{4} \times \frac{6}{{10}} = \frac{{11}}{{20}},
\end{eqnarray*}

     y por la regla de la suma (i.e., $p(F=m)+p(F=f)=1$), la probabilidad total de seleccionar una fresa es:

\begin{eqnarray*}
p(F=f) &=& 1 - p(F=m)  \\
               & = & 1 - \frac{{11}}{{20}} = \frac{9}{20}.
\end{eqnarray*}

3. Las probabilidades condicionales de seleccionar un color de caja dado un tipo de fruta:

*   Suponga que se ha seleccionado una fresa y se desea conocer de cuál caja fue tomada:
\begin{eqnarray*}
p(C = r|F = f)  &  = & \frac{{p(F = f| C = r)p(C = r)}}{{p(F = f)}}, \\
               & = & \frac{3}{4} \times \frac{4}{{10}} \times \frac{{20}}{9} = \frac{2}{3},
\end{eqnarray*}
y por la regla de la suma se tiene que $p(C = a|F = f) = 1 - 2/3 = 1/3$, lo cual indica que es más probable tomar una fresa que se encuentre en la caja roja.

*   Del mismo modo, suponga que se ha seleccionado una manzana y se desea conocer de cuál caja fue tomada:
\begin{eqnarray*}
p(C = r|F = m)  &  = & \frac{{p(F = m| C = r)p(C = r)}}{{p(F = m)}}, \\
               & = & \frac{1}{4} \times \frac{4}{{10}} \times \frac{{20}}{11} = \frac{2}{11},
\end{eqnarray*}
y por la regla de la suma se tiene que $p(C= a|F = m) = 1 - 2/11 = 9/11$, lo cual indica que es más probable tomar una manzana que se encuentre en la caja azul.


# **Ejemplo 2**

Una compañía de transporte público tiene tres líneas en una ciudad, de forma que el 45\% de los autobuses cubre el servicio de la línea 1, el 25\% cubre la línea 2 y el 30\% cubre el servicio de la línea 3. Se sabe que la probabilidad de que un autobús se averíe diariamente es del 2\%, 3\% y 1\%  para las líneas 1, 2, y 3, respectivamente.


La avería en un autobús  es una variable aleatoria denotada por $A$, la cual puede tomar los valores 1 (hay una avería) ó 0 (no hay una avería). De manera similar, la línea de autobús es una variable aleatoria denotada por $L$, la cual puede tomar los valores 1, 2 ó 3, para denotar a cada una de las rutas:

*   Probabilidad de servicio de la línea: $p(L=1)=9/20$, $p(L=2)=1/4$, y $p(L=3)=3/10$.
*   Probabilidad de sufrir una avería en la línea: $p(A=1|L=1)=1/50$, $p(A=1|L=2)=3/100$, y $p(A=1|L=3)=1/100$.

Entonces, a partir de la información dada por la compañía se pueden inferir las siguientes probabilidades:

1. La probabilidad total de que un autobús sufra una avería:

\begin{eqnarray*}
p(A=1)  &  = & \sum\limits_i {p(L = i,A = 1)}, \\
               & = & \sum\limits_i {p(A = 1|L = i)p(L = i)}, \\
               & = & \frac{1}{{50}} \times \frac{9}{{20}} + \frac{3}{{100}} \times \frac{1}{4} + \frac{1}{{100}} \times \frac{3}{{10}} = 0.0195 \nonumber
\end{eqnarray*}

2. La probabilidad de que un autobús no sufra una avería:
\begin{equation*}
p(A=0) = 1 - p(A=1) = 0.9805
\end{equation*}

3. ¿En qué línea de transporte es más probable que un autobús sufra una avería?

\begin{equation*}
p(L = i|A = 1) = \frac{{p(A = 1|L = i)p(L = i)}}{{p(A)}}.
\end{equation*}

Entonces para cada línea se tiene:

\begin{equation*}
p(L = 1|A = 1) = \frac{{0.02 \times 0.45}}{{0.0195}} = 0.4615.
\end{equation*}
\begin{equation*}
p(L = 2|A = 1) = \frac{{0.03 \times 0.25}}{{0.0195}} = 0.3846.
\end{equation*}
\begin{equation*}
p(L = 3|A = 1) = \frac{{0.01 \times 0.30}}{{0.0195}} = 0.1538.
\end{equation*}

Estos resultados indican que es más probable que la línea 1 sufra una avería.
