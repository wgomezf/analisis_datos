```python
import numpy as np
import matplotlib.pyplot as plt

def tlc_dados(n, m):
    # Simula el lanzamiento de n dados m veces
    x = np.random.randint(1, 7, size=(n, m))
    # Suma las caras de los dados
    y = np.sum(x, axis=0)
    # Grafica un histograma de frecuencias
    h, edges = np.histogram(y, bins=150)
    c = (edges[:-1] + edges[1:]) / 2
    xx = np.round(c)
    l = np.round(np.linspace(np.min(xx), np.max(xx), 6))
    fig = plt.figure(facecolor=(1, 1, 1), figsize=(7, 2.5))
    ax = fig.add_subplot(111)
    ax.bar(c, h,  color='blue', edgecolor='k')
    ax.set_xlabel(f'Suma de las caras de {n} dados lanzados {m} veces', fontsize=10)
    ax.set_ylabel('Frecuencia', fontsize=10)
    plt.show()
```

```python
tlc_dados(1, 1000)
```
